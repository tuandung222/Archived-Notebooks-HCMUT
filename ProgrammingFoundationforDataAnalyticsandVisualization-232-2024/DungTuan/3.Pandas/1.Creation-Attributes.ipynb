## 3.1- Dataset: ***sklearn.datasets.load_iris***
* References:
  * Link: [https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)

* Information:
  * ```X```: a Dataframe, contains all samples of the dataset.
    * rows: sample/observations/data points
    * cols: features/variables/attributes
  * ```y```: a series, **contains the label for the corresponding sample, i.e., the sample has the same index in ```y```**, for examples,
    * Sample X.iloc[0] has label is y[0]
    * Sample X.iloc[1] has label is y[1]
    * and so on

* Requirements:
  * Show the information of the dataset as in the output cell.
  

````python
from sklearn.datasets import load_iris

blob = load_iris(as_frame= True)
X, y = blob.data, blob.target
print("type(X): ", type(X))
print("type(y): ", type(y))

print()
print('='*100)

#YOUR CODE IS HERE

print("Dataset: IRIS")
print("-"*100)
print("Number of samples, len(x): ", len(X))
print("Number of features, X.shape[1]: ", X.shape[1])
print("Name of features, list(X.columns):")
print(list(X.columns))
print()
print("Columns's data type, X.dtypes")
print(X.dtypes)
print("Data type of label, y.dtype: ", y.dtype)
print()
print("-"*50)
print("Information of X, X.info(): ")
print(X.info())
print()
print("-"*50)
print("Description of X, X.describe().T: ")
print(X.describe().T)
````

### Question-1's requirements
* Information:
  * **files**: Python's list, contains all file files in the dataset
  * **labels**: Python's list, contains the label for the corresponding file in **files**, **corresponding** means **at the same index**

* Requurements:
  1. Create a DataFrame with columns are the data in ```files``` and ```labels```
  2. Show the results as in the output cell

* Guideline:
  * Use ```df.sample(10)``` to show 10 sampled randomly from the created dataframe

````python
from pathlib import Path

path_to_data = './bbc'
# List all files
all_paths = Path(path_to_data).rglob('*.txt')
files = [str(path) for path in all_paths]
labels = [file.split('/')[-2] for file in files]
print("files: ", files)
print("labels:", labels)
print("="*100)

#YOUR CODE IS HERE
df = pd.DataFrame(list(zip(files, labels)), columns=['files', 'labels'])
print("DataFrame showing files and their labels:")
print("df.sample(10): ")
print("-"*50)
print(df.sample(10))
print("="*100)
````
