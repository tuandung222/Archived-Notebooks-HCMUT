#1 Numbers in Python

##1.1 Samples for integer
The following example shows how to write an ***integer*** in Python.

Notes:
* ***type(.)***: return the variable's data type
* ***isinstance(variable,data-type)***: verify if the *variable* is an instance of the *data-type*
* ***str(.)***: convert to string type


```python
x = 10
print("x = {}".format(x))
print("Type of x: {}".format(str(type(x))))
print("x is an instance of int: {}".format(isinstance(x, int)))
print("x is an instance of float: {}".format(isinstance(x, float)))
print("x is an instance of complex: {}".format(isinstance(x, complex)))
```

#2 String in Python

##2.1 Samples

###2.1.1 How to write a string

```python
s1 = 'Programming foundation for data analytics and visualization'
s2 = "Programming foundation for data analytics and visualization"
s3 = """
    Course: Programming foundation for data analytics and visualization
    Field: Data Science
    """
print("Single-line string: ", s1)
print("Single-line string: ", s2)
print("Multi-lines string: ", s3)
```

###2.1.2 String formatted

```python
a, b = 10.5, 20.25
s = f'a = {a}, b = {b}'
print(s)
```

```python
a, b = 10.5, 20.25
s = 'a = {0}, b = {1}'.format(a,b) # a is at 0, b is at 1 in the parameter list
print(s)
```

```python
a, b = 10.5, 20.25
s1 = 'a = {:5.2f}\nb = {:10.4f}\n'.format(a,b)
s2 = 'a = {:5.2f}\nb = {:<15.4f}\n'.format(a,b) # <: left-alignment
s3 = 'a = {:5.2f}\nb = {:>15.4f}\n'.format(a,b) # >: right-alignment
s4 = 'a = {:5.2f}\nb = {:^15.4f}\n'.format(a,b) # ^: center-alignment
s5 = 'a = {:5.2f}\nb = {:015.4f}\n'.format(a,b) # 0 in 015.4f => preceeding with 0
s6 = 'b = {1:5.2f}\na = {0:015.4f}\n'.format(a,b) # 0 in 015.4f => preceeding with 0
print(s1)
print(s2)
print(s3)
print(s4)
print(s5)
print(s6)
```

###2.1.3 String formatted with dynamic width and precision

```python
a, b = 10.5, 20.1
line = 'a = {:0{w1}.{p1}f}; b={:0{w2}.{p2}f}'.format(a, b, w1=10, p1=2, w2=15, p2=5)
print(line)
```

###2.1.4 String repeated

```python
s1 = '#'*4 # repeat character # 4 times
s2 = '-'*10 # repeat character - 10 times
print(s1)
print(s2)
```
