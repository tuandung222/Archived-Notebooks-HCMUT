## Q.2 - Defining/Visulizing 3D-Mesh using **np.meshgrid**
* Requirements:
  * Create a 3D-mesh as shown in the figure below
* Guidelines:
  * See the example: [https://matplotlib.org/stable/gallery/mplot3d/surface3d.html](https://matplotlib.org/stable/gallery/mplot3d/surface3d.html)

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

#YOUR CODE IS HERE

x = np.linspace(-4, 4 + 0.1, 500)
y = np.linspace(-4, 4 + 0.1, 500)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

cm = plt.get_cmap("coolwarm")
fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, Z, cmap="coolwarm", linewidth=0, antialiased=False)

ax.set_zlim(0, 50.0)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter("{x:.02f}")
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

# Section-5: Sampling normal distribution using **np.random.normal**

## Examples

```python
from scipy.stats import norm

# Declare variables
mu, sigma = 5, 3
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = norm.pdf(x, mu, sigma)


plt.figure(figsize=(6,4))
plt.plot(x, y, c='r', linewidth=2)
title = f"Normal distribution with mu={mu}, sigma={sigma}"
plt.title(title)
plt.show()
```

## Q.1 Sampling numbers from normal distribution with given **mean** and **std**
* Variables:
  * **mu**, **sigma**: are numbers defining mean and standard deviation of normal distribution

  * **N**: integer, number of samples
* Requirements:
  * Generate **N** numbers from the normal distribution
    * NOTE: **Each time you run the code different integers are generated**
  * Show the distribution as the figure below
*Guidelines:
  * Link to wiki::normal distribution: [https://en.wikipedia.org/wiki/Normal_distribution](https://en.wikipedia.org/wiki/Normal_distribution)
  * Link to numpy API: [https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.htmlt](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
# Declare variables
mu, sigma = 5, 3
N = [100, 1000, 10000, 100000, 1000000]

#YOUR CODE IS HERE
# show hist, sublot size: 3x2
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
# set title for the whole figure
fig.suptitle("Histograms from normal distribution with mu=5, sigma=3")
# show only 5 histograms
for i, n in enumerate(N):
    row, col = i // 2, i % 2
    X = np.random.normal(mu, sigma, n)
    ax = axs[row, col]
    ax.hist(X, bins=100, density=True)
    ax.set_title(f"Histogram from {n} points")
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
    y = norm.pdf(x, mu, sigma)
    ax.plot(x, y, c='r', linewidth=2)
plt.tight_layout()
fig.delaxes(axs[-1, -1])
plt.show()
```

## Q.2- Sampling 2D-normal

### Q.2.1 Examples

```python
# Declare variables
mu = (5, 3)
sigma = (3,4)
N = 10000

# Generate data from 2D-normal(mu sigma)
X = np.random.normal(mu, sigma, (N,2)) #N rows, 2 cols

# Show the 2D-histogram
fig = plt.figure(figsize=(6,4))
ax = fig.gca() # you can get axes from plt.subplot(,,)
_, _, _, im = ax.hist2d(X[:, 0], X[:, 1], bins=10*10, density=True)
title = "Histogram from {} points".format(N)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title(title)
fig.colorbar(im, ax=ax)
plt.show()
```
