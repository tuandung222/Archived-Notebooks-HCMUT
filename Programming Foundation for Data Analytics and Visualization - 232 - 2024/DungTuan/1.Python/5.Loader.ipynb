```python
import os
import glob
import functools as F
import random

import os
import glob
import functools as F
import random

class DataLoader(object):
  def __init__(self, path: str, batch_size=5):
    # YOUR CODE IS HERE
    self.batch_size = batch_size
    self.path = path
    self.prepare()

  def prepare(self):
    # YOUR CODE IS HERE
    self.classes = [
      x
      for x in os.listdir(self.path)
      if os.path.isdir(os.path.join(self.path, x))
    ]
    self.list_files_labels = [
      (x, y)
      for y in self.classes
      for x in glob.glob(os.path.join(self.path, y, "*.jpg"))
    ]
    self.current_index = 0

  def __iter__(self):
    # YOUR CODE IS HERE
    self.current_index = 0
    self.shuffle_items = random.sample(
      self.list_files_labels, len(self.list_files_labels)
    )
    return self

  def __next__(self):
    # YOUR CODE IS HERE
    if self.current_index >= len(self.shuffle_items):
      raise StopIteration
    return_tuples = self.shuffle_items[
      self.current_index : self.current_index + self.batch_size
    ]
    self.current_index = self.current_index + self.batch_size
    return [x[0] for x in return_tuples], [x[1] for x in return_tuples]
```

```python
loader = DataLoader("./animals/")
for idx, (files, labels) in enumerate(loader):
  print(f"batch: {idx}")
  print("files: ", files)
  print("labels: ", labels)
```

```python
"""
load_data: implement it as a generator
"""
def load_data(path_to_data:str, batch_size=5):
    # YOUR CODE IS HERE
    loader = DataLoader(path_to_data, batch_size)
    for idx, (files, labels) in enumerate(loader):
        yield files, labels



```

```python
loader = load_data("./animals/")
for files, labels in loader:
  print(files)
  print(labels)
```
